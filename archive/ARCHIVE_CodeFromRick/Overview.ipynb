# Overview

There are different situations we should consider:

- We know the market share of each individual EPD -- then it's straightforward to calculate the weighted average (but we don't know the market shares, in general)
- We know the market share of subgroups of products, and we know which group each EPD belongs to, but not the relative shares within each group -- this can be split into two subgroups, where we first need to find the representative distribution for the subgroup (which may fall in any of these cases, e.g. the subgroup-representative GWP may be known or not), and then it's like case 1 where we know the market share of the subgroups -> weight directly
- We know the market share of subgroups of products, but we **don't know** for sure which group each EPD belongs to --> set the cluster market shares directly (rather than inferring them) but use the inference to assign EPDs to clusters
- We don't know the market shares, but we do know what the average GWP result should be
